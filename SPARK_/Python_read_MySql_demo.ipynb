```python

# python3 
# https://dluo.me/s3databoto3

import os 
import pandas as pd
import pymysql
from pymysql import * 
```

```python

def get_mysql_conn(host, user, database):
        #conn = pymysql.connect(
        #host = '127.0.0.1',
        #port = 3306,
        #user = 'root',
        #db = 'local_dev')
        conn = pymysql.connect(
                host = host,
                port = 3306,
                user = user,
                db =database)
        return conn 
```

```python
host = '127.0.0.1'
user = 'root'
database = 'local_dev'
mysql_conn = get_mysql_conn(host, user, database)
cursor = mysql_conn.cursor()
```

```python
print (mysql_conn)
print (cursor)
```

```python
# read data via pymysql

SQL = "select * from movie_metadata limit 10 ; "
cursor.execute(SQL)
cursor.fetchall()
```

```python
# read data via pandas

df = pd.read_sql(SQL,con=mysql_conn)
df
```
